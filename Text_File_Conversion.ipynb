```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[200,300,60,800,1000]
C=[12,20,34,45,55]
CPP=[15,25,36,50,60]
Java=[18,30,40,55,70]
R=[20,35,45,60,80]
Python=[25,40,53,75,100]
picture = sns.lineplot(x=x,y=C,label='C')
sns.lineplot(x=x,y=CPP,label='C++')
sns.lineplot(x=x,y=Java,label='Java')
sns.lineplot(x=x,y=R,label='R')
sns.lineplot(x=x,y=Python,label='Python')
picture.set_title('Comparison of Performance in Loading the Files')
picture.set_ylabel('File Size (in MB)')
picture.set_xlabel('Time Taken (in sec)')
```
